```python
import networkx as nx
import math
import matplotlib.pyplot as plt
```

```python
#added to match nodes
def nmatch(n1, n2):
    return n1==n2
```

```python
#red bird graph
G3 = nx.DiGraph()
G3.add_node("bird", id = "bird")
G3.add_node("red", id = "red")
G3.add_node("shape_bird", id = "shape_bird")
G3.add_edge("bird","red")
G3.add_edge("bird","shape_bird")

pos = nx.spectral_layout(G3)

nx.draw(G3, pos, node_size=1000, alpha=0.9,
    labels={node: node for node in G3.nodes()})
nx.draw_networkx_edge_labels(
    G3, 
    pos,
    edge_labels={('bird', 'red'): 'colour', 
                 ('bird', 'shape_bird'): 'shape'}
)
plt.axis('off')
plt.show()
```

```python
#blue bird graph
G4 = nx.DiGraph()
G4.add_node("bird", id = "bird")
G4.add_node("blue", id = "blue")
G4.add_node("shape_bird", id = "shape_bird")
G4.add_edge("bird","blue")
G4.add_edge("bird","shape_bird")
pos = nx.spectral_layout(G4)

nx.draw(G4, pos, node_size=1000, alpha=0.9,
    labels={node: node for node in G4.nodes()})
nx.draw_networkx_edge_labels(
    G4, 
    pos,
    edge_labels={('bird', 'blue'): 'colour', 
                 ('bird', 'shape_bird'): 'shape'}
)
plt.axis('off')
plt.show()
```

```python
nx.graph_edit_distance(G3,G4, node_match=nmatch)
```
